### Mathematical 
The kineteic energy of the horizontal beam of electons is

$$KE = \frac{1}{2}mv^2_x = eV_1$$

Although, the voltages used lead to slightly relativistic speeds 

$$KE = (\gamma -1)mc^2$$

where 

$$\gamma=\sqrt{\frac{1}{1-\frac{v^2_x}{c^2}}}$$

We use the classical description and compare the results to those using relativistic calculations.

This gives

$$v_x = \sqrt{\frac{2eV_1}{m}}$$

Next, we consider the deflection. The force applied by $V_2$ is

$$F = eE = e\frac{V_2}{d} = ma$$

And the acceleration gives a velocity along the y-direction

$$v_y = v_{oy} + at$$

where the initial velocity $v_{oy}=0$. From this, we can write

$$v_y = \frac{e}{m}\frac{V_2}{d}t$$

The time $t$ that the electron stays between the deflection plates is $t=\ell/v_x$. Thus,

$$v_y = \frac{e}{m}\frac{V_2\ell}{v_xd}$$

The deflection will look like Figure 3. 
<img src=./deflection.png alt="image of cathode ray deflection" width="500">
<img src=./deflection_angle.png alt="image of cathode ray deflection measurement" width="500">
<center>Figure 3. Deflection of the electron beam by perpendicular voltage.</center>

From the geometry of the deflection, we deduce that

$$\tan\theta = \frac{e}{m}\frac{\ell}{d}\frac{1}{v^2_x}V_2$$

A graph of $\tan\theta$ vs. $V_2$ will be linear with a slope of

$${\rm slope} = \frac{e\ell}{mdv^2_x}$$

J.J. Thomson was able to determine the horizontal velocity $v_x$ by balancing deflecting electric and magnetic forces equal and opposite to one another.

$$F_{net} = e\vec{E} + e\vec{v}\times \vec{B} = 0$$

and if $\vec{v} \perp \vec{B}$ with $v$ along $x$, $E$ along $-z$, and $B$ along $y$

$$eE_{-z} + ev_xB_y = 0$$

By deflecting the horizontal electron beam with an electric field and then, undoing the deflection with a magnetic field, one obtains

$$v_x = \left|\frac{E}{B}\right|$$

We cannot do this with our apparatus. So, to calculate $v_x$ we must use $e/m$ and pretend like we didn't.

### $V_1 = 3000$ Volts
Let's look at the first set of data Table 1.

```python
# Raw Data
e = 1.602e-19
me = 9.11e-31
x = 0.01 #beam manipulating plate diameter
d = 0.002#beam manipulating plate separation
V_1 = 3000#voltage that accelerates the beam of electrons to v_x
v_x = np.sqrt(2*e*V_1/me)#math to calculate the initial horizonatal electron velocity v_x
V_2 = np.array([50,100,150,200]) #independent variable settings
theta = np.array([0.0417, 0.0833, 0.125, 0.167]) #dependent variable measurements
theta_unc = np.array([0.005, 0.005, 0.005, 0.005])
#linear fit function
def f_line(x, m, b):
    return m*x + b

#Fit the line to your data
#NOTE: You will need to do everything below for each accelerating voltage, V1
#Copy it into new code cells for each. 
#Be sure to insert markdown cells to tell the reader what each code cell is for.
m = 0.008#initial guess at the slope
b = 1#initial guess at the intercept
fit_params, fit_cov = curve_fit(f_line, V_2, theta, (m,b), sigma=theta_unc, absolute_sigma=True)
#fit_params = [slope, intercept]
#fit_cov = uncertainties squared
print(fit_params, fit_cov)

#Create a fit array from the fit_params
#slope = fit_params[0]
#intercept = fit_params[1]
fit_y = fit_params[0]*V_2+fit_params[1]#m*x + b

#Graph the results by filling in all the appropriate details
plt.errorbar(V_2, theta, yerr=theta_unc, fmt='or', capsize=5.0, capthick=1.0, label='data')#plot the data
plt.plot(V_2, fit_y, '-k', label='fit')#plot the fit
plt.xlabel(r'$V_2$')
plt.ylabel(r'$\theta$ (rad)')
plt.legend()
plt.show()
```

Wow, that looks nice. The value of $e/m$ from the slope is

$$\begin{align}
{\rm slope} &= \frac{e\ell}{mdv^2_x}\\
\frac{e}{m} &= \frac{{\rm slope}~dv^2_x}{\ell}\\
\frac{e}{m} &= 1.76\times 10^{-11}{\rm C/kg}
\end{align}$$

The product/quotient rule gives an uncertainty of

$$\Delta \left(\frac{e}{m}\right) = \left(\frac{\Delta {\rm slope}}{\rm slope} + \frac{\Delta d}{d} + \frac{\Delta \ell}{\ell} + \frac{2\Delta v_x}{v_x}\right)\frac{e}{m}$$

and from the derivative rule

$$\Delta v_x = \frac{\sqrt{2e/mV_1}}{v_x}\Delta V_1$$

```python
em = fit_params[0]*d*v_x**2/x
em_unc = (np.sqrt(fit_cov[0,0])/fit_params[0] + 0.0001/d + 0.0001/x + 2*np.sqrt(2*e/(me*V_1))/v_x*50)*em
print('e/m= {:.3e}'.format(em), '+/- {:.3e}'.format(em_unc))
```

## Conclusions
We expect a value of  $e/m=1.72\times 10^{11}$ C/kg. Our experiments give values of

$$\frac{e}{m} = 1.762\pm 0.259\times10^{11} {\rm C/kg}$$

and

$$\frac{e}{m} = 1.768\pm 0.239\times10^{11} {\rm C/kg}$$

Both values agree with the expected value within the experimental uncertainty. However, the uncertainties are quite large (around 10% relative uncertainty). We do have unaccounted for relativistic effects. The velocities $v_x$ are greater than 10% the speed of light from classical calculations. The error of using a classical calculation is made worse for the higher accelerating voltage, which may have lead to a less accurate result.

### How does relativistic correction change the results?
Let's take a look at using relativistic kinetic energy. Rearranging the voltage-KE equation, we get

$$\gamma = \frac{eV_1}{mc^2}+1$$

We can solve for $\gamma$ and then solve $\gamma$ for speed.

$$v_x = \sqrt{\left(1-\frac{1}{\gamma^2}\right)c^2}$$

A plot of $v_x$ is shown below comparing classical and relativistic calculations for the range of voltages $V_1$ used in this experiment.

```python
e = 1.602e-19
m = 9.11e-31
c = 3e8
volt = np.linspace(2500, 5000, 50)
v_class = np.sqrt(e*volt/(0.5*m))
gam = e*volt/(m*c**2)+1
v_rel = np.sqrt((1-1/gam**2)*c**2)#np.sqrt(1-(m*c**2/(e*volt+m*c**2))**2*c**2)

plt.plot(volt, v_class, '-r', label="classical")
plt.plot(volt, v_rel, '--b', label="relativistic")
plt.xlabel('Acc Volts')
plt.ylabel('velocity')
plt.legend()
plt.show()
```

### Recalculating the speed to get $e/m$

We will recalculate $\gamma$ and $v_x$ for each accelerating voltage. These are used to recalculate $e/m$ from the slopes above. There is no change in the those fit parameters.

```python
# Recalculate the horizontal speed
gamma = e*V_1/(m*c**2)+1
v_x = np.sqrt((1-1/gamma**2)*c**2)#math to calculate the initial horizonatal electron velocity v_x

#print(gamma, v_x2)
em_rel = fit_params[0]*d*v_x**2/x
em_rel_unc = (np.sqrt(fit_cov[0,0])/fit_params[0] + 0.0001/d + 0.0001/x + 2*np.sqrt(2*e/(me*V_1))/v_x*50)*em_rel
print('e/m= {:.3e}'.format(em_rel), '+/- {:.3e}'.format(em_rel_unc))

# Recalculate the horizontal speed
gamma = e*V_12/(m*c**2)+1
v_x2 = np.sqrt((1-1/gamma**2)*c**2)#math to calculate the initial horizonatal electron velocity v_x

#print(gamma, v_x2)
em2_rel = fit_params2[0]*d*v_x2**2/x
em2_rel_unc = (np.sqrt(fit_cov2[0,0])/fit_params2[0] + 0.0001/d + 0.0001/x + 2*np.sqrt(2*e/(me*V_12))/v_x2*50)*em2_rel
print('e/m= {:.3e}'.format(em2_rel), '+/- {:.3e}'.format(em2_rel_unc))
```

This brings the results closer to the expected value.
